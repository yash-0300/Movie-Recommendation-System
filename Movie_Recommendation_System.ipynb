```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
print(movies.shape)
movies.head(3)
```

```python
movies.info()
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(3)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
get_genres = movies.iloc[0].genres
print(get_genres)
print(type(get_genres))
```

```python
import ast
def convertGenresKeywords(obj):
    lst = []
    for i in ast.literal_eval(obj):
        lst.append(i['name'])
    return lst
```

```python
movies['genres'] = movies['genres'].apply(convertGenresKeywords)
```

```python
movies.head(3)
```

```python
movies['keywords'] = movies['keywords'].apply(convertGenresKeywords)
```

```python
import ast
def convertCast(obj):
    lst = []
    counter = 0
    for i in ast.literal_eval(obj):
        if (counter != 3):
          lst.append(i['name'])
        else:
          break
        counter = counter + 1
    return lst
```

```python
movies['cast'] = movies['cast'].apply(convertCast)
```

```python
movies['crew'][0]
```

```python
def convertCrew(obj):
    lst = []
    for i in ast.literal_eval(obj):
        if (i['job'] == 'Director'):
          lst.append(i['name'])
          break
    return lst
```

```python
movies['crew'] = movies['crew'].apply(convertCrew)
```

```python
movies.head(3)
```

```python
movies['overview'] = movies['overview'].apply(lambda string: string.split())
```

```python
movies.head(3)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ", "") for i in x])

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(3)
```

```python
movies_df = movies[['movie_id', 'title', 'tags']]
movies_df.head(3)
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:" ".join(x))
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
movies_df.head(3)
```

```python
!pip install nltk
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizeText(text):
  lst = []
  for i in text.split():
    lst.append(lemmatizer.lemmatize(i))
  return " ".join(lst)
```

```python
movies_df['tags'] = movies_df['tags'].apply(lemmatizeText)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = vectorizer.fit_transform(movies_df['tags']).toarray()
print(vectors.shape)
```

```python
vectors[0]
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
# Function will recommend the top 5 movies based on the given movie
def RecommendMovies(movie):
    # Fetch index of movie in the dataframe
    movie_index = movies_df[movies_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
```

```python
RecommendMovies('Batman Begins')
```
